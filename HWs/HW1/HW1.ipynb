<div dir="rtl">
داده‌های گمشده (Data missing) زمانی رخ می‌دهد که متغیری در یک مشاهده مقدار داده‌ای نداشته باشد. مثلا در همین دیتافریم در هر سطر اگر مقدار ستونی NaN یا Null باشد یعنی آن ستون در آن مشاهده مقدار ندارد و Data Missing رخ داده است. <br>
روش‌های متداول برای حل این مشکل عبارت اند از:
    <ul>
        <li>جانهی(imputation):
            در این روش داده‌هایی که مقدار داده‌ای ندارند، مقداردهی می‌شوند. یکی از راه‌ها برای انجام این کار درون‌یابی است که در داده‌های عددی مورد استفاده قرار می‌گیرد. کاربرد این روش در این مجموعه داده می‌تواند در ستون‌هایی باشد که داده عددی دارند و در بعضی سطر‌ها مقدار ندارند. 
        </li>
        <br>
        <li>حذف جزئی(partial deletion):
            در این روش داده‌ها را به داده‌هایی که داده گمشده ندارند محدود می‌کنیم. برای این کار نیز دو روش وجود دارد:<br>
            ۱) حذف لیستی (listwise deletion) که در آن سطری که یک یا چند ستون آن مقدار داده‌ای نداشته باشند  را حذف می‌کنیم. این روش تعداد داده‌ها را کم می‌کند ولی داده‌هایی که باقی می‌مانند هیچ ستون خالی‌ای ندارند و اطلاعات آن‌ها کامل است. کاربرد این روش در این داده‌ها برای حالت‌هایی است که بعضی ستون‌ها در تعداد کمی از داده‌ها مقدار ندارند پس با حذف کل ستون مقدار زیادی از اطلاهات مفید از بین می‌رود پس به جای حذف ستون کافیست سطر‌هایی که ستون در آن‌ها مقدار ندارد راحذف کنیم تا اطلاعات کمتری از دست برود. <br>
            ۲) حذف روجی (pairwise deletion) که در آن ستون‌هایی که در بعضی از سطرها مقدار داده‌ای ندارند را حذف می‌کنیم. این روش به ما اجازه می‌دهد از همه داده‌ها استفاده کنیم ولی اطلاعات ما از آن داده را کم می‌کند چون ستون‌هایی از جدول را حذف می‌کند. کاربرد این روش در این داده‌ها برای ستون‌هایی است که در تعداد زیادی از سطر‌ها مقدار ندارند، مثلا ستون‌هایی که در بیشتر از نصف سطر‌ها مقدار نداشته باشند.
        </li>
        <br>
        <li>تحلیل کامل(full analysis): در این روش از همه اطلاعات موجود بدون در نظر گرفتن داده‌های بی‌مقدار استفاده می‌شود که خود شامل دو رویکرد افتراقی و مولد است. کاربرد این روش در این داده‌ها به این صورت است که هنگام انجام محاسبات داده‌های بی‌مقدار را در نظر نگیریم. در این حالت هیچ داده‌ای از دست نمی‌رود ولی محاسبات دشوار می‌شود.
        </li>
        <br>
        <li>روش‌های مبتنی بر مدل(model based techniques):
        روش‌های مبتنی بر مدل با استفاده از نمودار، ابزارهای دیگری برای تست داده‌های گم‌شده و  برآورد پارامترها تحت شرایطی که داده‌های گم‌شده وجود دارند پیشنهاد می‌دهند.
        </li>
    </ul>
        
</div>

<div dir="rtl">
تا اینجای کار فقط داشتیم داده‌هامون رو تمیز می‌کردیم. از اینجا به بعد می‌خوایم کمی روی اونها کار کنیم. یکی از 
فاکتورهای مهم توی کار کردن میزان حقوقه. پس بیاید و ستون تخمین حقوق رو به دو ستون تبدیل کنید که حداقل و حداکثر حقوق 
درخواستی در اون درج شده.
سپس ستون میانگین حقوق را از این دو ستون جدید به دست بیارید و در انتها پنج داده‌ی اول مجموعه داده رو نمایش بدید.
    (۸ امتیاز)
</div>

```python
def min_salary(salary):
    parts = salary.split("-") 
    min_salary = int(re.search(r'\d+', parts[0].strip()).group())
    return min_salary

def max_salary(salary):
    parts = salary.split("-")
    max_salary = int(re.search(r'\d+', parts[1].strip()).group())
    return max_salary

df['Min_Salary'] = df['Salary Estimate'].apply(min_salary)
df['Max_Salary'] = df['Salary Estimate'].apply(max_salary)
df['Mean_Salary'] = (df.Min_Salary + df.Max_Salary)/2
df.head(5)
```

<div dir="rtl">
    می‌خوایم یه آمار کلی از حقوق کارها داشته باشیم. کاری که می‌خوایم بکنید اینه که ۵٪ از کارها با بالاترین میانگین حقوق و ۵٪ از کارها با پایین‌ترین میانگین حقوق رو حذف کنید و بعدش یه آمار کلی از میانگین حقوق‌ها بدید و همچنین بگید که بنظرتون چرا ۵٪ از بالا و ۵٪ از پایین رو حذف کردیم؟ آمار کلی‌ای که می‌خوایم شامل:  mean, std, min, max می‌شه.
    (۶ امتیاز)
</div>

<div dir="rtl">
با حذف ۵ درصد بالا و پایین در واقع داریم داده‌های پرت را از بین می‌بریم تا نتایج دقیق تری به دست بیاید.
</div>

```python
jobs = df.groupby(["Job Title"]).mean()
n = int(5 * jobs.shape[0] / 100)
large = jobs.nlargest(n, 'Mean_Salary')
small = jobs.nsmallest(n, 'Mean_Salary')
pd.concat([jobs, small]).drop_duplicates(keep=False, inplace=True)
pd.concat([jobs, large]).drop_duplicates(keep=False, inplace=True)
jobs.describe()[["Mean_Salary"]]
```

```python
def clean_company_name(c_name):
    if '\n' in c_name:
        c_name = c_name[0: c_name.index('\n')]
    return c_name


df['Company Name'] = df['Company Name'].apply(clean_company_name)
df_comp = df.groupby(["Company Name", "Job Title"]).mean()
df_comp = df_comp.groupby(["Company Name"]).mean().reset_index()
# df_comp[["Company Name","Mean_Salary"]].head(5)
df_comp[["Mean_Salary"]].plot.kde()
```

```python
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
state_geo = f'{url}/us-states.json'
my_map_2 = folium.Map(location=[48, -102], zoom_start=3)
data = df_3.groupby(["state"]).count()[["Job Title"]].reset_index()
folium.Choropleth(
    geo_data=state_geo,
    name='choropleth',
    data=data,
    columns=['state', 'Job Title'],
    key_on='feature.id',
    fill_color='BuPu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Job count',
    nan_fill_color='grey',
    nan_opacity="0.5"
).add_to(my_map_2)
folium.LayerControl().add_to(my_map_2)
my_map_2
```
